## Предобработка данных

### Предобработка датасета матча резюме и вакансий

```python
match_df = pd.read_csv('cv2vacancy.tsv', sep='\t')
print(match_df.info())
match_df.head()
```

```python
match_df = match_df[['id', 'idCv', 'idVacancy']].drop_duplicates()
match_df.head()
```

### Предобработка датасета вакансий

```python
vacancy_df = pd.read_csv('vacancy.tsv', sep='\t', dtype=str)
print(vacancy_df.info())
vacancy_df.head()
```

Выделяем столбцы, которые могут быть полезны, при сопоставлении вакансии и резюме.

```python
vacancy_df = vacancy_df[['id', 'vacancy_name', 'busy_type', 'schedule_type', 'position_responsibilities',
                         'position_requirements', 'education', 'education_speciality',
                         'languageKnowledge', 'hardSkills', 'softSkills']]
vacancy_df.head()
```

Фильтруем вакансии, чтобы рассматривать только подмножетсво IT вакансий.

```python
positions = []
keywords = ['программист', 'разработчик', 'ml', 'аналитик', 'тестировщик',
            'devops', 'mlops', 'data', 'фронтенд', 'бэкенд', 'python', 'java', 'javascript', 'c++', 
            'c#', 'ruby', 'swift', 'objective-c', 'go', 'php', 'kotlin', 'typescript']
for position in pd.unique(vacancy_df['vacancy_name']):
    if type(position) is not str: 
        print('NOT STR', position)
        continue
    if any(keyword in position.lower() for keyword in keywords):
        print(position)
        positions.append(position)
```

```python
new_vacancy_df = vacancy_df[vacancy_df['vacancy_name'].isin(positions) & vacancy_df['id'].isin(match_df['idVacancy'])]
new_vacancy_df.info()
```

Переименовываем столбцы датасета, чтобы объединить все датасеты в один. 

```python
new_vacancy_df = new_vacancy_df.rename(columns={
    'id': 'idVacancy',
    'vacancy_name': 'vacancy_positionName',
    'busy_type': 'vacancy_busyType',
    'schedule_type': 'vacancy_scheduleType',
    'position_responsibilities': 'vacancy_responsibilities',
    'position_requirements': 'vacancy_requirements',
    'education': 'vacancy_education',
    'education_speciality': 'vacancy_educationSpeciality',
    'languageKnowledge': 'vacancy_languageKnowledge',
    'hardSkills': 'vacancy_hardSkills', 
    'softSkills': 'vacancy_softSkills'
}).reset_index(drop=True)
new_vacancy_df.head()
```

Преобразуем данные в столбцах `vacancy_languageKnowledge`, `vacancy_hardSkills`, `vacancy_softSkills` в более удобный для обработки формат.

```python
for i, text in enumerate(new_vacancy_df["vacancy_languageKnowledge"]):
    language_text = ''
    for language in json.loads(text):
        if 'code_language' in language:
            language_text += language['code_language']
        if 'level' in language:
            language_text += f": {language['level']}"
        language_text += "\n"
    new_vacancy_df.iloc[i]["vacancy_languageKnowledge"] = language_text[:-1]
new_vacancy_df.head(20)
```

```python
for i, text in enumerate(new_vacancy_df["vacancy_hardSkills"]):
    skills_text = ''
    for skill in json.loads(text):
        if 'hard_skill_name' in skill:
            skills_text += f"{skill['hard_skill_name']}, "
    new_vacancy_df.iloc[i]["vacancy_hardSkills"] = skills_text[:-2]
new_vacancy_df.tail(20)
```

```python
for i, text in enumerate(new_vacancy_df["vacancy_softSkills"]):
    skills_text = ''
    for skill in json.loads(text):
        if 'soft_skill_name' in skill:
            skills_text += f"{skill['soft_skill_name']}, "
    new_vacancy_df.iloc[i]["vacancy_softSkills"] = skills_text[:-2]
new_vacancy_df.head(20)
```

Объединяем датасеты матча и вакансий в один

```python
merged_df = pd.merge(match_df, new_vacancy_df, how="inner", on=['idVacancy'])
print(merged_df.info())
merged_df.head()
```

### Предобработка датасета резюме

```python
cv_df = pd.read_csv('cv.tsv', sep='\t')
print(cv_df.info())
cv_df.head()
```

Выделяем столбцы, которые могут быть полезны, при сопоставлении вакансии и резюме.

```python
cv_df = cv_df[['id', 'positionName', 'experience', 'educationList', 'additionalEducationList',
               'workExperienceList', 'hardSkills', 'softSkills', 'scheduleType', 'busyType', 'languageKnowledge']]
cv_df = cv_df[cv_df['id'].isin(merged_df['idCv'])].reset_index(drop=True)
print(cv_df.info())
cv_df.head()
```

Преобразуем данные в столбцах `educationList`, `additionalEducationList`, `workExperienceList`, `languageKnowledge`, `hardSkills`, `softSkills` в более удобный для обработки формат.

```python
pattern = re.compile(r'"(\w+)":\s*(".*?(?<!\\)"|\d+|\w+)', re.UNICODE)

for i, text in enumerate(cv_df["educationList"]):
    edu_text = ''
    for edu in text[2:-2].split("} {"):
        edu = {key: value for key, value in pattern.findall(edu)}
        if 'instituteName' in edu:
            edu_text += f"{edu['instituteName'][1:-1]}\n"
        if 'graduateYear' in edu:
            edu_text += f"Год окончания: {edu['graduateYear']}\n"
        if 'level' in edu:
            edu_text += f"Уровень образования: {edu['level'][1:-1]}\n"
        if 'faculty' in edu:
            edu_text += f"Факультет: {edu['faculty'][1:-1]}\n"
        if 'specialty' in edu:
            edu_text += f"Специальность: {edu['specialty'][1:-1]}\n"
        if 'qualification' in edu:
            edu_text += f"Квалификация: {edu['qualification'][1:-1]}\n"
        if len(edu_text) > 0:
            edu_text += "\n"
    cv_df.loc[i, "educationList"] = edu_text
cv_df.head(20)
```

```python
for i, text in enumerate(cv_df["additionalEducationList"]):
    edu_text = ''
    for edu in text[2:-2].split("} {"):
        edu = {key: value for key, value in pattern.findall(edu)}
        if 'courseName' in edu:
            edu_text += f"{edu['courseName'][1:-1]}\n"
        if 'organization' in edu:
            edu_text += f"{edu['organization'][1:-1]}\n"
        if 'graduateYear' in edu:
            edu_text += f"Год окончания: {edu['graduateYear']}\n"
        if len(edu_text) > 0:
            edu_text += "\n"
    cv_df.loc[i, "additionalEducationList"] = edu_text
cv_df.head(20)
```

```python
for i, text in enumerate(cv_df["workExperienceList"]):
    work_text = ''
    for work in text[2:-2].split("} {"):
        work = {key: value for key, value in pattern.findall(work)}
        if 'companyName' in work:
            work_text += f"{work['companyName'][1:-1]}, "
        if 'jobTitle' in work:
            work_text += f"{work['jobTitle'][1:-1]}\n"
        if 'demands' in work:
            work_text += f"Обязанности: {work['demands']}\n"
        if 'archievements' in work:
            work_text += f"Достижения: {work['archievements']}\n"
        if len(work_text) > 0:
            work_text += "\n"
    cv_df.loc[i, "workExperienceList"] = work_text
cv_df.head(20)
```

```python
for i, text in enumerate(cv_df["hardSkills"]):
    skills_text = ''
    for skill in text[2:-2].split("} {"):
        skill = {key: value for key, value in pattern.findall(skill)}
        if 'hardSkillName' in skill:
            skills_text += f"{skill['hardSkillName'][1:-1]}, "
    cv_df.loc[i, "hardSkills"] = skills_text[:-2]
cv_df.head(20)
```

```python
for i, text in enumerate(cv_df["softSkills"]):
    skills_text = ''
    for skill in text[2:-2].split("} {"):
        skill = {key: value for key, value in pattern.findall(skill)}
        if 'softSkillName' in skill:
            skills_text += f"{skill['softSkillName'][1:-1]}, "
    cv_df.loc[i, "softSkills"] = skills_text[:-2]
cv_df.head(20)
```

```python
for i, text in enumerate(cv_df["languageKnowledge"]):
    language_text = ''
    for language in text[2:-2].split("} {"):
        language = {key: value for key, value in pattern.findall(language)}
        if 'codeLanguage' in language:
            language_text += language['codeLanguage'][1:-1]
        if 'level' in language:
            language_text += f": {language['level'][1:-1]}"
        language_text += "\n"
    cv_df.loc[i, "languageKnowledge"] = language_text[:-1]
cv_df.head(20)
```

Переименовываем столбцы датасета, чтобы объединить все датасеты в один.

```python
cv_df = cv_df.rename(columns={
    'id': 'idCv',
    'positionName': 'cv_positionName',
    'scheduleType': 'cv_scheduleType',
    'busyType': 'cv_busyType',
    'experience': 'cv_experience',
    'educationList': 'cv_educationList',
    'additionalEducationList': 'cv_additionalEducationList',
    'workExperienceList': 'cv_workExperienceList',
    'languageKnowledge': 'cv_languageKnowledge',
    'hardSkills': 'cv_hardSkills', 
    'softSkills': 'cv_softSkills'
})
cv_df.head()
```

Объединяем все в один.

```python
merged_df = pd.merge(merged_df, cv_df, how="inner", on=['idCv'])
print(merged_df.info())
merged_df.head()
```
