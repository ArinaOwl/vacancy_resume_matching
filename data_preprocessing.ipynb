## Разметка датасета

```python
df = pd.read_csv('cv2vacancy_merged_clean.tsv', sep='\t')
df.head()
```

```python
new_df = df[["idCv", "idVacancy"]]
```

Объединим информацию в блоки выбранные эмпирически: `Название позиции`, `Требования/опыт`, `Образование`, `Навыки`.

```python
def sum_texts(cols):
    res = []
    for i in range(len(cols[0])):
        new_res = ''
        for col in cols:
            if str(col[i]) != 'nan':
                new_res += f"{str(col[i])}\n\n"
        res.append(new_res)
    return res
```

```python
new_df.loc[:, "vacancy_positionName"] = df["vacancy_positionName"]
new_df["vacancy_busyAndScheduleType"] = sum_texts([
    df["vacancy_busyType"],
    df["vacancy_scheduleType"]
])
new_df.loc[:, "vacancy_responsibilitiesAndRequirements"] = sum_texts([
    df["vacancy_responsibilities"],
    df["vacancy_requirements"]
])
new_df.loc[:, "vacancy_education"] = sum_texts([
    df["vacancy_education"],
    df["vacancy_educationSpeciality"]
])
new_df.loc[:, "vacancy_skills"] = sum_texts([
    df["vacancy_hardSkills"],
    df["vacancy_softSkills"],
    df["vacancy_languageKnowledge"]
])
new_df.head()
```

```python
new_df.loc[:, "cv_positionName"] = df["cv_positionName"]
new_df["cv_busyAndScheduleType"] = sum_texts([
    df["cv_busyType"],
    df["cv_scheduleType"]
])
new_df.loc[:, "cv_experience"] = sum_texts([
    df["cv_experience"],
    df["cv_workExperienceList"]
])
new_df.loc[:, "cv_education"] = sum_texts([
    df["cv_educationList"],
    df["cv_additionalEducationList"]
])
new_df.loc[:, "cv_skills"] = sum_texts([
    df["cv_hardSkills"],
    df["cv_softSkills"],
    df["cv_languageKnowledge"]
])

new_df.head()
```

Для каждого блока **описания вакансии** векторизуем полученный текст с помощью BERT модели `cointegrated/rubert-tiny`.

```python
import torch
from transformers import AutoTokenizer, AutoModel
tokenizer = AutoTokenizer.from_pretrained("cointegrated/rubert-tiny")
model = AutoModel.from_pretrained("cointegrated/rubert-tiny")
# model.cuda()  # uncomment it if you have a GPU

def embed_bert_cls(text, model, tokenizer):
    t = tokenizer(text, padding=True, truncation=True, return_tensors='pt')
    with torch.no_grad():
        model_output = model(**{k: v.to(model.device) for k, v in t.items()})
    embeddings = model_output.last_hidden_state[:, 0, :]
    embeddings = torch.nn.functional.normalize(embeddings)
    return embeddings[0].cpu()#.numpy()
```

```python
vacancy_df = new_df[["idVacancy", "vacancy_positionName",
                     "vacancy_busyAndScheduleType", "vacancy_responsibilitiesAndRequirements",
                     "vacancy_education", "vacancy_skills"]].drop_duplicates().reset_index(drop=True)
vacancy_df.info()
```

```python
vacancy_positionName_embeds = np.array([])
vacancy_busyAndScheduleType_embeds = np.array([])
vacancy_responsibilitiesAndRequirements_embeds = np.array([])
vacancy_education_embeds = np.array([])
vacancy_skills_embeds = np.array([])


for i in tqdm(range(len(vacancy_df))):
    vacancy_positionName_embeds = np.append(vacancy_positionName_embeds,
                                            embed_bert_cls(vacancy_df.loc[i, "vacancy_positionName"], model, tokenizer).numpy())
    vacancy_busyAndScheduleType_embeds = np.append(vacancy_busyAndScheduleType_embeds,
                                                   embed_bert_cls(vacancy_df.loc[i, "vacancy_busyAndScheduleType"], model, tokenizer).numpy())
    vacancy_responsibilitiesAndRequirements_embeds = np.append(vacancy_responsibilitiesAndRequirements_embeds,
                                                               embed_bert_cls(vacancy_df.loc[i, "vacancy_responsibilitiesAndRequirements"], model, tokenizer).numpy())
    vacancy_education_embeds = np.append(vacancy_education_embeds,
                                         embed_bert_cls(vacancy_df.loc[i, "vacancy_education"], model, tokenizer).numpy())
    vacancy_skills_embeds = np.append(vacancy_skills_embeds,
                                      embed_bert_cls(vacancy_df.loc[i, "vacancy_skills"], model, tokenizer).numpy())
```

Рассчитаем косинусную схожесть между всеми описаниями вакансий в датасете. Схожесть будем рассчитывать для каждого блока отдельно, а затем вычислим взвешенное среднее.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vacancy_positionName_cos_sims = cosine_similarity(vacancy_positionName_embeds.reshape(1892, -1), vacancy_positionName_embeds.reshape(1892, -1))
print(vacancy_positionName_cos_sims, '\n')
print(f"Min: {np.min(vacancy_positionName_cos_sims)}")
print(f"Max: {np.max(vacancy_positionName_cos_sims)}")
print(f"Mean: {np.mean(vacancy_positionName_cos_sims)}")
```

```python
vacancy_busyAndScheduleType_cos_sims = cosine_similarity(vacancy_busyAndScheduleType_embeds.reshape(1892, -1), vacancy_busyAndScheduleType_embeds.reshape(1892, -1))
print(vacancy_busyAndScheduleType_cos_sims, '\n')
print(f"Min: {np.min(vacancy_busyAndScheduleType_cos_sims)}")
print(f"Max: {np.max(vacancy_busyAndScheduleType_cos_sims)}")
print(f"Mean: {np.mean(vacancy_busyAndScheduleType_cos_sims)}")
```

```python
vacancy_responsibilitiesAndRequirements_cos_sims = cosine_similarity(vacancy_responsibilitiesAndRequirements_embeds.reshape(1892, -1),
                                                                     vacancy_responsibilitiesAndRequirements_embeds.reshape(1892, -1))
print(vacancy_responsibilitiesAndRequirements_cos_sims, '\n')
print(f"Min: {np.min(vacancy_responsibilitiesAndRequirements_cos_sims)}")
print(f"Max: {np.max(vacancy_responsibilitiesAndRequirements_cos_sims)}")
print(f"Mean: {np.mean(vacancy_responsibilitiesAndRequirements_cos_sims)}")
```

```python
vacancy_education_cos_sims = cosine_similarity(vacancy_education_embeds.reshape(1892, -1),
                                               vacancy_education_embeds.reshape(1892, -1))
print(vacancy_education_cos_sims, '\n')
print(f"Min: {np.min(vacancy_education_cos_sims)}")
print(f"Max: {np.max(vacancy_education_cos_sims)}")
print(f"Mean: {np.mean(vacancy_education_cos_sims)}")
```

```python
vacancy_skills_cos_sims = cosine_similarity(vacancy_skills_embeds.reshape(1892, -1),
                                            vacancy_skills_embeds.reshape(1892, -1))
print(vacancy_skills_cos_sims, '\n')
print(f"Min: {np.min(vacancy_skills_cos_sims)}")
print(f"Max: {np.max(vacancy_skills_cos_sims)}")
print(f"Mean: {np.mean(vacancy_skills_cos_sims)}")
```

```python
vacancy_cos_sims_agr = (
    2 * vacancy_positionName_cos_sims +
    1 * vacancy_busyAndScheduleType_cos_sims +
    7 * vacancy_responsibilitiesAndRequirements_cos_sims +
    3 * vacancy_education_cos_sims +
    7 * vacancy_skills_cos_sims
) / 20


print(f"Min: {np.min(vacancy_cos_sims_agr)}")
print(f"Max: {np.max(vacancy_cos_sims_agr)}")
print(f"Mean: {np.mean(vacancy_cos_sims_agr)}")

for i in range(vacancy_cos_sims_agr.shape[0]):
    for j in range(vacancy_cos_sims_agr.shape[1]):
        if i > j:
            vacancy_cos_sims_agr[i,j] = 1
print(vacancy_cos_sims_agr)
```

Выделим менее схожие по косинусной мере (< 0.5) вакансии, чтобы составить датасет отрицательных примеров матча вакансии и резюме.

```python
neg_dict = dict()
for i in range(vacancy_cos_sims_agr.shape[0]):
    for j in range(i + 1, vacancy_cos_sims_agr.shape[1]):
        if vacancy_cos_sims_agr[i, j] < 0.5:
            if i not in neg_dict:
                neg_dict[i] = [j]
            else:
                neg_dict[i].append(j)
print(neg_dict)
```

```python
df_pos_targets = df[["idCv", "idVacancy"]]
df_pos_targets["target"] = np.ones(len(df_pos_targets))
df_pos_targets
```

```python
df_neg_targets = pd.DataFrame(columns=df_pos_targets.columns)
for key, item in tqdm(neg_dict.items()):
    vacancy_ids_for_item = vacancy_df.loc[item, "idVacancy"]
    id_cv = df[df["idVacancy"].isin(vacancy_ids_for_item)]["idCv"].to_list()

    id_vacancy = [vacancy_df.loc[key, "idVacancy"] for _ in range(len(id_cv))]

    target = np.zeros(len(id_cv))

    new_df_neg_targets = pd.DataFrame(list(zip(id_cv, id_vacancy, target)), columns=df_pos_targets.columns)

    df_neg_targets = pd.concat([df_neg_targets, new_df_neg_targets], ignore_index = True)
df_neg_targets.info()
```

```python
df_targets = pd.concat([df_neg_targets, df_pos_targets], ignore_index = True)
df_targets.info()
```

Итоговый датасет содержит размеченные данные положительных и отрицательных примеров матча для обучения модели бинарной классификации.
